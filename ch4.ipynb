```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
N = 500 # Total number of flips
pHeads = 0.5 # Probability of head
```

```python
# simulate the flipping experiment
flip_sequence = [random.choice([0,1]) for i in range(N)]
```

```python
# compute the proportion at each experiment
head_prop = [np.sum(flip_sequence[:i]) / i for i in range(1,N+1)]
```

```python
# Translate the beginning sequence
init_seq = ''.join(['H' if entry == 0 else 'T' for entry in flip_sequence[0:10]])
```

```python
# Make the plot
plt.plot(range(N), head_prop)
plt.scatter(range(N), head_prop, alpha=0.7)
plt.plot(range(N), [0.5 for i in range(N)], linestyle=":")
plt.xscale('log') # log x axis following the original figure
plt.ylim(0, 1)
plt.xlabel('Flip number')
plt.ylabel('Proportion heads')
plt.text(8, 0.8, 'Flip sequence: ' + init_seq + '...')
plt.text(8, 0.7, 'End Proportion: ' + str(head_prop[-1]))
```
